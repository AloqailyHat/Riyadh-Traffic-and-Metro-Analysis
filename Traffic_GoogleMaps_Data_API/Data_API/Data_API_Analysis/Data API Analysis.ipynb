## Google API
This is a live traffic from 9/9/2024 5:00 PM to 10/9/2024 5:00 PM for interchange station

```python
# This code is the extracted live from google maps API where we extracted the traffic times around 6, 7, 8, 9, 10, 11 AM, 12, 1, 2, 3, 4, 5, 6, 7 PM
# origin: from
# destination: to

import pandas as pd

# Load your dataset
df = pd.read_csv('lines_cal.csv')

# Convert "Request Time" to datetime to extract the hour
df['Request Time'] = pd.to_datetime(df['Request Time'], format='%H:%M')

# Extract hour from "Request Time"
df['Hour'] = df['Request Time'].dt.hour

# Define a function to convert duration to minutes
def duration_to_minutes(duration):
    if 'hour' in duration:
        parts = duration.split(' ')
        hours = int(parts[0]) * 60
        minutes = int(parts[2])
        return hours + minutes
    elif 'mins' in duration:
        return int(duration.split(' ')[0])
    return 0

# Apply the function to convert "Duration" to minutes
df['Avg Duration in Minutes'] = df['Duration'].apply(duration_to_minutes)

# Apply the function to convert "Duration with Traffic" to minutes
df['Avg Duration with Traffic in Minutes'] = df['Duration with Traffic'].apply(duration_to_minutes)

# Filter the data for the time range between 6 AM and 7 PM (6 to 19 in 24-hour format)
filtered_data = df[(df['Hour'] >= 6) & (df['Hour'] <= 19)]

# Function to convert 24-hour time to 12-hour format with AM/PM
def convert_to_ampm(hour):
    if hour == 0:
        return "12 AM"
    elif hour < 12:
        return f"{hour} AM"
    elif hour == 12:
        return "12 PM"
    else:
        return f"{hour - 12} PM"

# Apply the AM/PM conversion
filtered_data['Hour (AM/PM)'] = filtered_data['Hour'].apply(convert_to_ampm)

# Calculate the average duration and duration with traffic per hour for each Origin, Destination, and Distance
df_lines = filtered_data.groupby(['Origin', 'Destination', 'Hour (AM/PM)', 'Distance'])[
    ['Avg Duration in Minutes', 'Avg Duration with Traffic in Minutes']
].mean().reset_index()

# Print or save the result
df_lines
```
